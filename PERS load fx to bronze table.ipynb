# Load fx data into bronze table


```python

# Mount external lakehouse  "PERS" into specified local path.
mssparkutils.fs.mount("abfss://c36724bb-dd17-4c25-9f59-85e9014e1a0a@onelake.dfs.fabric.microsoft.com/f5a2609b-9d1b-4f99-8cbc-e3181f4b3bbd", "/lakehouse/PERS")

import pandas as pd
# Load data into pandas DataFrame from mssparkutils.fs.getMountPath('/lakehouse/PERS/') + "Files/fxdata.csv"
df = pd.read_csv(mssparkutils.fs.getMountPath('/lakehouse/PERS/') + "Files/fxdata.csv", skiprows=1, skipfooter=1, header=0, engine='python')
display(df)
```

```python
# Convert to spark df

spark_df = spark.createDataFrame(df)
```

```python
# Use format and save to load as a Delta table
table_name = "bronze_fxdata"
spark_df.write.mode("overwrite").format("delta").save(f"Tables/{table_name}")

# Confirm load as Delta table
print(f"Spark DataFrame saved to Delta table: {table_name}")
```
